#### Import Data

```python
# Load the CSV
df= pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/edx/project/drinks.csv')
df.head()
```

```python
df.rename(columns={'country':'Country', 'beer_servings':'Beer Servings', 'spirit_servings':'Spirit Servings', 'wine_servings':'Wine Servings', 
                   'continent':'Continent', 'total_litres_of_pure_alcohol':'Pure Alcohol'}, inplace=True)
df.head()
```

<b> Question 6:</b> Fit a linear regression model to predict the <code>'total_litres_of_pure_alcohol'</code> using the number of <code>'wine_servings'</code> then calculate $R^{2}$:

```python
# Create LR Object
lm=LinearRegression()

# Predict total Pure Alcohol using Wine Servings
X=df[['Wine Servings']]
Y=df['Pure Alcohol']

# Fit the model and calculate R-square
lm.fit(X,Y)
print("R-square value:", lm.score(X,Y))
```

<b> Extra Question 2:</b> Fit a linear regression model to predict the <code>'Pure Alcohol'</code> using the number of <code>'Beer Servings'</code> then calculate $R^{2}$:

```python
# Predict total Pure Alcohol using Beer Servings
X=df[['Beer Servings']]
y=df['Pure Alcohol']

# Fit the model and calculate R-square
lm.fit(X,y)
print("R-square value:", lm.score(X,y))
```

<b> Question 7:</b> Use the list of features to predict the <code>'total_litres_of_pure_alcohol'</code>, split the data into training and testing and determine the $R^2$ on the test data, using the provided code:

```python
# Define Variables
variables=['Beer Servings', 'Wine Servings', 'Spirit Servings']
X=df[variables]

# Define test train variables 
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.1, random_state=0)

# Fit and Score Data
lm.fit(X_train,y_train)
print("Test Score:\n""R-square value:", lm.score(X_test,y_test))
```

<b>Question 8 :</b> Create a pipeline object that scales the data, performs a polynomial transform and fits a linear regression model. Fit the object using the training data in the question above, then calculate the R^2 using. the test data. Take a screenshot of your code and the $R^{2}$. There are some hints in the notebook:

```python
# Create pipe input
Input=[('scale',StandardScaler()), ('polynomial', PolynomialFeatures(include_bias=False)), ('model', LinearRegression())]
pipe=Pipeline(Input)

# Pipe fit&score
pipe.fit(X_train,y_train)
print('Pipe Score:',pipe.score(X_test,y_test))
```

<b>Question 9</b>: Create and fit a Ridge regression object using the training data, setting the regularization parameter to 0.1 and calculate the $R^{2}$ using the test data. Take a screenshot of your code and the $R^{2}$

```python
RR=Ridge(alpha=0.1)

# Ridge Regression fit&score
RR.fit(X_train,y_train)
print('Ridge Regression Score:',RR.score(X_test, y_test))
```
