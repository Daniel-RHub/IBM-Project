<b> Question 6:</b> Fit a linear regression model to predict the <code>'total_litres_of_pure_alcohol'</code> using the number of <code>'wine_servings'</code> then calculate $R^{2}$:

```python
# Create LR Model
lm=LinearRegression()

# Variables
X=df[['Wine Servings']]
y=df['Pure Alcohol']

# Fit&Score
lm.fit(X,y)
print("R-Square Score:",lm.score(X,y))
```

<b> Question 7:</b> Use the list of features to predict the <code>'total_litres_of_pure_alcohol'</code>, split the data into training and testing and determine the $R^2$ on the test data, using the provided code:

```python
# Create Variable
var=['Beer Servings', 'Wine Servings', 'Spirit Servings']
X=df[var]

# Train&Test
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.1, random_state=0)

# Fit&Score
lm.fit(X_train,y_train)
print("Train Score\n""R-square Score:",lm.score(X_test,y_test))
```
